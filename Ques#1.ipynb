```python
#Author: Farhan Shakil
#Course: Python Programming Language 
#Roll # PY01052
#Email: mfarhanshakil@gmail.com
#Assignment # 3
#Question # 1

while True:
    print("\t\tCALCULATOR")
    print("\n Enter \"A\" or \"a\" to Perform Addition")
    print(" Enter \"S\" or \"s\" to Perform Subtraction")
    print(" Enter \"M\" or \"m\" to Perform Multiplication")
    print(" Enter \"D\" or \"d\" to Perform Division")
    print(" Enter \"P\" or \"p\" to Perform Power Function")

    choice = input("\n Enter Your Choice ===>>>   ")

    while choice != 'A' and choice != 'a' and choice != 'S' and choice != 's' and choice != 'M' and choice != 'm' and choice != 'D' and choice != 'd' and choice != 'P' and choice != 'p':
        print("\n INVALID CHOICE")
        choice= input(" Enter Your Choice Again from Given Menu ===>>>   ")
    
    first_number  = float(input("\n Enter First Number  ===>>>   "))
    second_number = float(input(" Enter Second Number ===>>>   "))

    if choice == 'A' or  choice == 'a':
        print("\n\t ADDITION OPERATION")
        print(first_number , "+" , second_number , "=" , first_number + second_number)
    
    elif choice == 'S' or  choice == 's':
        print("\n\t SUBTRACTION OPERATION")
        print(first_number , "-" , second_number , "=" , first_number - second_number)
    
    elif choice == 'M' or  choice == 'm':
        print("\n\t MULTIPLICATION OPERATION")
        print(first_number , "x" , second_number , "=" , first_number * second_number)
    
    elif choice == 'D' or  choice == 'd':
        print("\n\t DIVISION OPERATION")
        print(first_number , "รท" , second_number , "=" , first_number / second_number)
    
    elif choice == 'P' or  choice == 'p':
        print("\n\t POWER OPERATION")
        print(first_number , "^" , second_number , "=" , first_number ** second_number)
        
    cont_prog = input("\n Press Enter to Perform Another Opertion or Enter \"E\" or \'e\' to Exit ===>>> ")
    
    if cont_prog == 'E' or cont_prog == 'e':
        break
```
